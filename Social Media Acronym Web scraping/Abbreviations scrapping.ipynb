```python
import requests
from bs4 import BeautifulSoup

url = "https://kutola.co.uk/blog/100-social-media-acronyms-and-abbreviations-you-need-to-know/"

# Step1: Get the HTML code in our python program using GET method

r = requests.get(url)
htmlContent = r.content
# This will show us the raw HTML content

#print(htmlContent)
r.close()
```

```python
# Step2: Parse the Tree traversal
soup = BeautifulSoup(htmlContent, 'html.parser')
```

```python
print(soup.title)
```

```python
required_div = soup.find('div', {"class":"post-entry-content clrfix"})
```

```python
print(required_div)
```

```python
abrvs = required_div.find_all('li')
```

```python
print(abrvs)
```

```python
print(abrvs[78].text)
```

```python
abb = abrvs[0].text
```

```python
abb_list = abb.split(':')
```

```python
from collections import defaultdict
```

```python
for abrv in abrvs:
    print(abrv.text.lower())
```

```python
list_of_abrvs = []
for abrv in abrvs:
    list_of_abrvs.append(abrv.text.lower().split(':'))
    
list_of_abrvs
```

```python
ordict = defaultdict(list)
for key, val in list_of_abrvs:
    ordict[key].append(val)
```

```python
ordict
```

```python
import json
```

```python
with open("Abbreviations_lowered.json", "w") as outfile:
    json.dump(ordict, outfile)
```

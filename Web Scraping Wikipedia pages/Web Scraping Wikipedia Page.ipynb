```python
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"

# Step1: Get the HTML code in our python program using GET method

r = requests.get(url)
htmlContent = r.content
# This will show us the raw HTML content

#print(htmlContent)
r.close()
```

```python
# Step2: Parse the Tree traversal
soup = BeautifulSoup(htmlContent, 'html.parser')
#print(soup.prettify) # print the parsed data of html
```

```python
# Lets get all the links in the page along with its attributes such as href, title and inner text
for link in soup.find_all('a'):
    print('id: {}'.format(link.get('id')))
    print('class: {}'.format(link.get('class')))
    print('href: {}'.format(link.get('href')))
    print('Title: {}'.format(link.get('title')))
    print('Inner Text: {}'.format(link.text))
    print('-------------*----------------*--------------------*----------------*---------------\n')
```
